```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Analyses of Philadephia and Global average temperature data 

```python
#Import data with Pandas
city_data = pd.read_csv('cdata.csv')
glob_data = pd.read_csv('gdata.csv')

city_data = city_data.dropna() # droped the missing values
yr = city_data['year']
Yr= glob_data['year']

city_avg_temp = city_data['avg_temp']
gl_avg_temp = glob_data['avg_temp']
```

```python
#using Matplotlib.pyplot to plot the line chart
plt.title('Line Chart')
plt.xlabel('Year')
plt.ylabel('Average Temp')
plt.plot(year,city_avg_temp, label ="Phila avg_temp")
plt.plot(Yr,gl_avg_temp, label = "Global")
plt.legend()
plt.grid()
plt.show()
```

Similarity: 1) There is a constant increase in temperature in the 2000s. 2) Before 1800, the city and the world experience cooler temperature.
Differnce:(1)The temperature in Philadephia is hotter than the global temperature from 1800s. 
    (2) In the year 1743 and 1779, the temperature in philadephia is cooler than the global temperature.
    
            
            

```python
#7 days moving average
# To calculate the moving average, we use the rolling function. This helps in the moving window count.
# Here window length = 7
city = city_avg_temp.rolling(7) # 
glob = gl_avg_temp.rolling(7)

# calculate the mean of each rolling window
mov_avg = city.mean() 
mov_avg1 = glob.mean()
```

# Plot of the Moving Average

```python

plt.title('7 Days Moving Average')
plt.xlabel('Year')
plt.ylabel('Moving Average Temp')
plt.plot(year,mov_avg, label ="Phila avg_temp")
plt.plot(Yr,mov_avg1, label = "Global")
plt.legend()
plt.grid()
plt.show()
```

```python
#window length = 20
city = city_avg_temp.rolling(20)
glob = gl_avg_temp.rolling(20)

mov_avg = city.mean()
mov_avg1 = glob.mean()
```

```python

plt.title('20 Days Moving Average')
plt.xlabel('Year')
plt.ylabel('Moving Average Temp')
plt.plot(year,mov_avg, label ="Phila avg_temp")
plt.plot(Yr,mov_avg1, label = "Global")
plt.legend()
plt.grid()
plt.show()
```

The change in temperature of city temp and global temp are similar.
Both had change in temperature in the 1900s in which moving from cooler temperature to hotter temperature.
